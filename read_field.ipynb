```python
field_dict = {}
with open('field.tsv', 'r', encoding='utf-8') as file:
    header = file.readline().strip().split('\t')
    for line in file:
        parts = line.strip().split('\t')
        row_obj = dict(zip(header, parts))
        field_id = row_obj['field_id']
        field_dict[field_id] = row_obj
```

```python
def list_columns(field_id, custom_max_instance=None, custom_max_array=None):
    row = field_dict.get(str(field_id))
    
    # Generate column names
    columns = []

    isInstanced = int(row['instanced']) > 0
    isArray = int(row['arrayed']) > 0

    instance_min = int(row['instance_min'])
    instance_max = int(row['instance_max'])
    array_min = int(row['array_min'])
    array_max = int(row['array_max'])

    if custom_max_instance is not None:
        instance_max = custom_max_instance
    if custom_max_array is not None:
        array_max = custom_max_array

    for i in range(instance_min, instance_max+1):
        for j in range(array_min, array_max+1):
            col_name = f'p{field_id}'
            if isInstanced:
                col_name += f'_i{i}'
            if isArray:
                col_name += f'_a{j}'
            columns.append(col_name)

    return columns
```
